```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
data = np.loadtxt('scan.dat')
```

```python
f = data[:,0]
tau = data[:,1]
Neff = data[:, 2]
positive_mask = Neff > 0
min_positive = Neff[positive_mask].min() if np.any(positive_mask) else 1e-3
Neff_safe = np.where(Neff > 0, Neff, min_positive)
```

```python
plt.tricontourf(f, tau, Neff_safe, levels=500, norm = LogNorm())  # 100 contour levels for smoothness
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$f_{DDM}$')
plt.ylabel(r'$\tau$ (years)')
plt.title(r'$N_{eff}$')
plt.colorbar(label=r'$N_{eff}$')
plt.axhline(378000, color='k', linestyle='--')
plt.tight_layout()
```
