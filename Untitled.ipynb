```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
data = np.loadtxt('scan.dat')
```

```python
f = data[:,0]
tau = data[:,1]
atr = data[:, 2]
Neff = data[:, 3]
```

```python
plt.tricontourf(f, tau, Neff, norm = LogNorm())
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$f_{DDM}$')
plt.ylabel(r'$\tau$ (years)')
plt.title(r'$N_{eff}$')
plt.axhline(378000, color='k', linestyle='--')
plt.tight_layout()
```

```python
plt.tricontourf(f, atr, Neff, norm = LogNorm())
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$f_{DDM}$')
plt.ylabel(r'$a_{tr}$')
plt.title(r'$N_{eff}$')
plt.axhline(9e-4, color='k', linestyle='--')
plt.tight_layout()
```
