```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_atr = np.loadtxt('rhoDMatr.dat')
data_tau = np.loadtxt('rhoDMtau.dat')
```

```python
fatr = data_atr[:,0]
tauatr = data_atr[:,1]
Neffatr = data_atr[:,2]
Omegadiffatr = data_atr[:, 3]
ftau = data_tau[:,0]
atrtau = data_tau[:,1]
Nefftau = data_tau[:, 2]
Omegadifftau = data_tau[:, 3]
```

```python
fig1, (Ntau, difftau) = plt.subplots(1,2, figsize = (10,6))
Ntau.tricontourf(ftau, atrtau, Nefftau, levels = 10)
Ntau.set_xscale('log')
Ntau.set_yscale('log')
Ntau.set_xlabel(r'$f_{DDM}$')
Ntau.set_ylabel(r'$\tau$ (years)')
Ntau.set_title(r'$N_{eff}$, $a_{tr} = 1/(1+1100)$')
Ntau.axhline(378000, color='k', linestyle='--')
fig1.colorbar(Ntau.tricontourf(ftau, atrtau, Nefftau, levels = 10), ax=Ntau,
                     orientation='horizontal').set_label(r'$\Delta N_{eff}$')

difftau.tricontourf(ftau, atrtau, Omegadifftau, levels = 10)
difftau.set_xscale('log')
difftau.set_yscale('log')
difftau.set_xlabel(r'$f_{DDM}$')
difftau.set_ylabel(r'$\tau$ (years)')
difftau.set_title(r'$\Omega_{\mathrm{decay}} - \Omega_{\mathrm{observed}}$, $a_{tr} = 1/(1+1100)$')
difftau.axhline(378000, color='k', linestyle='--')
fig1.colorbar(difftau.tricontourf(ftau, atrtau, Omegadifftau, levels = 10), ax=difftau,
                     orientation='horizontal').set_label(r'$\Omega_{\mathrm{decay}} - \Omega_{\mathrm{observed}}$')

fig1.tight_layout()
```

```python
fig2, (Natr, diffatr) = plt.subplots(1,2, figsize = (10,6))
Natr.tricontourf(fatr, tauatr, Neffatr, levels = 10)
Natr.set_xscale('log')
Natr.set_yscale('log')
Natr.set_xlabel(r'$f_{DDM}$')
Natr.set_ylabel(r'$atr$')
Natr.set_title(r'$N_{eff}$, $\tau = 378,000$ years')
Natr.axhline(1/(1101), color='k', linestyle='--')
fig2.colorbar(Natr.tricontourf(fatr, tauatr, Neffatr, levels = 10), ax=Natr,
                     orientation='horizontal').set_label(r'$\Delta N_{eff}$')

diffatr.tricontourf(fatr, tauatr, Omegadiffatr, levels = 10)
diffatr.set_xscale('log')
diffatr.set_yscale('log')
diffatr.set_xlabel(r'$f_{DDM}$')
diffatr.set_ylabel(r'$atr$')
diffatr.set_title(r'$\Omega_{\mathrm{decay}} - \Omega_{\mathrm{observed}}$, $\tau = 378,000$ years')
diffatr.axhline(1/(1101), color='k', linestyle='--')
fig2.colorbar(diffatr.tricontourf(fatr, tauatr, Omegadiffatr, levels = 10), ax=diffatr,
                     orientation='horizontal').set_label(r'$\Omega_{\mathrm{decay}} - \Omega_{\mathrm{observed}}$')

fig2.tight_layout()
```
