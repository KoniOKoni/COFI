```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
data_atr = np.loadtxt('rhoDMatr.dat')
data_tau = np.loadtxt('rhoDMtau.dat')
```

```python
fatr = data_atr[:,0]
tauatr = data_atr[:,1]
Neffatr = data_atr[:,2]
ftau = data_tau[:,0]
atrtau = data_tau[:,1]
Nefftau = data_tau[:, 2]
```

```python
plt.tricontourf(ftau, atrtau, Nefftau, levels = 10)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$f_{DDM}$')
plt.ylabel(r'$\tau$ (years)')
plt.title(r'$N_{eff}$, $a_{tr} = 1/(1+1100)$')
plt.axhline(378000, color='k', linestyle='--')
plt.colorbar()
plt.tight_layout()
```
